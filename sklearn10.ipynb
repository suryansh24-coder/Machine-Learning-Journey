```python
# K-mean cluster :
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate data set :-
np.random.seed(42)
data = {
    "Annual Income" : np.random.randint(30000,100000,100) ,
    "Spending Score" : np.random.randint(1,100,100)
}
df = pd.DataFrame(data)
df.head()
```

```python
df
```

```python
plt.scatter(df['Annual Income'],df['Spending Score'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title("Customer Data - Annual Income vs Spending Score")
plt.show()
```

```python
X = df.values
kmeans = KMeans(n_clusters=3 , random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
print(df)
```

```python
plt.scatter(df['Annual Income'],df['Spending Score'],  c=df["Cluster"], cmap='rainbow' )
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title("Customer Data - Annual Income vs Spending Score")
plt.show()
```

```python
user_input = {
    "Annual Income" : [int(input("Enter your Annual Income : "))],
    "Spending Score" : [int(input("Enter your Spending Score : "))]
}
user_df = pd.DataFrame(user_input)
user_cluster = kmeans.predict(user_df)
print(user_cluster)
print(f"The user Belongs to cluster {user_cluster[0]}")
```
