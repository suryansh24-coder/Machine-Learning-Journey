```python
# Agglomerative Hierarchical Clustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
# Data Set :-
np.random.seed(42)
data = pd.DataFrame({
    "Age" : np.random.randint(18, 65, 100),
    "Income" : np.random.randint(20000, 100000, 100),
    "Spending Score" : np.random.randint(1, 100, 100)
    })
```

```python
scaler =  StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
model = AgglomerativeClustering(n_clusters=3 , metric='euclidean' ,linkage='ward') # euclidean show distance metrics , ward - it is a linkage method for joining 2 cluster groups
data['Cluster'] = model.fit_predict(scaled_data)
```

```python
# Visualization :-
plt.scatter(data['Income'], data['Spending Score'] ,c=data['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```
