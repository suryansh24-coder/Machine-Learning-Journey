```python
# GMM :  Gaussian Mixture Model :-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
# Dataset :-
np.random.seed(42)
segment1 = np.random.normal(loc=30 , scale = 5, size=100)
segment2 = np.random.normal(loc=60 , scale = 10, size=150)
segment3 = np.random.normal(loc=90 , scale = 5, size=120)
data = np.concatenate([segment1,segment2,segment3]).reshape(-1,1)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
n_components = 3
gmm = GaussianMixture(n_components=n_components, random_state=42)
gmm.fit(data_scaled)
```

```python
cluster_labels = gmm.predict(data_scaled)
```

```python
# Visualization
plt.scatter(data , np.zeros_like(data), c=cluster_labels, cmap='viridis')
plt.xlabel('Data')
plt.ylabel('Cluster')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
# User Testing
user_input =  float(input("Enter a purchase amount to predict the customer segment : "))
user_input_scaled = scaler.transform(np.array([[user_input]]))
prediction = gmm.predict(user_input_scaled.reshape(-1,1))[0]
print(f"The predicted customer segment for a purchase amount of {user_input} is : {prediction + 1}")
```
